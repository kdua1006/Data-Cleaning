```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('BL-Flickr-Images-Book.csv')
```

```python
data.head()
```

```python
#dropping not important columns
to_drop = ['Edition Statement',
           'Corporate Author', 
           'Corporate Contributors', 
           'Former owner', 
           'Engraver', 
           'Contributors',
           'Issuance type', 
           'Shelfmarks']
```

```python
data.drop(to_drop, inplace=True, axis=1)
#could also pass this as
#data.drop(columns=to_drop, inplace=True)
```

```python
#reexamining dataframe
data.head()
```

```python
#Let's change the index of the dataframe so that we can identify specific books by its Identifier
#First check if the Identifiers are unique
data['Identifier'].is_unique
```

```python
#Now, set new index
data = data.set_index('Identifier')
```

```python
#Further inspecting the data
data.dtypes.value_counts()
```

```python
data.head(10)
```

```python
#Removing all dates wiht extra characters and designating Publishing Date as the start date where valid
extract_dates = data['Date of Publication'].str.extract(r'^(\d{4})', expand=False)
```

```python
extract_dates.head(10)
```

```python
#Converting this to numeric
data['Date of Publication'] = pd.to_numeric(extract_dates)
```

```python
#Checking percentage of NaN and null values
data['Date of Publication'].isnull().sum()/len(data)

#~11%, not too shabby!
```

```python
#Cleaning the Place of Publication column
publication = data['Place of Publication']
```

```python
#Finding places of publication 
london = publication.str.contains('London')
oxford = publication.str.contains('Oxford')
```

```python
data['Place of Publication'] = np.where(london, 'London', np.where(oxford, 'Oxford', publication.str.replace('-', ' ')))
```
